## Data Visualisation

```python
df['State'].value_counts().plot(kind = 'bar')
```

```python
df['Vehicle_Category'].value_counts().plot(kind = 'pie', autopct ='%.2f%%')
```

```python
# Set plot style for better aesthetics
sns.set(style='whitegrid')

# Create a figure with specific size
plt.figure(figsize=(8, 5))

# Add a title to the plot
plt.title('Yearly Analysis of EV Sales in India', fontsize=14, fontweight='bold')

# Plot the line chart with markers
sns.lineplot(
    x='Year', 
    y='EV_Sales_Quantity', 
    data=df, 
    marker='o', 
    color='b', 
    linewidth=2
)

# Set labels for the axes with font size
plt.xlabel('Year', fontsize=12)
plt.ylabel('EV Sales Quantity', fontsize=12)

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style for the plot
sns.set(style='whitegrid')

# Sort data by total EV Sales Quantity in descending order
sorted_df = df.groupby('State', as_index=False, observed=False)['EV_Sales_Quantity'].sum().sort_values(by='EV_Sales_Quantity', ascending=False)

# Plot the sorted bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=sorted_df, 
    x='State', 
    y='EV_Sales_Quantity', 
    errorbar=None  # Disable confidence intervals
)

# Add title and labels
plt.title('EV Sales Quantity by State ', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('State', fontsize=14)
plt.ylabel('Total EV Sales Quantity', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Add horizontal gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better fit
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='State', hue='Vehicle_Category')
plt.title('Vehicle Category Distribution by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Vehicle Category')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(15,4))
sns.barplot(x='Vehicle_Class', y='EV_Sales_Quantity',data=df, hue='Vehicle_Class', palette='bright')
plt.title('Analysis by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('EV Sales')
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Vehicle_Category', y='EV_Sales_Quantity',data=df, hue='Vehicle_Category', palette='bright')
plt.title('Analysis by Vehicle Category')
plt.xlabel('Vehicle Category')
plt.ylabel('EV Sales')
plt.xticks(rotation=0);
```
